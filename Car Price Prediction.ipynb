# Categorical Features To Numeric Type

```python
car = pd.get_dummies(car,columns = ["fuel","seller_type","transmission"],drop_first=True)
car.drop(columns=['year'],inplace=True)
car.head()
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
np.random.seed(0)
car_train,car_test = train_test_split(car, test_size=0.3, random_state=100)
```

```python
car.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
var=["selling_price",'km_driven','Car_Age',"mileage","engine","max_power"]

car_train[var]=scaler.fit_transform(car_train[var])
car_test[var]=scaler.transform(car_test[var])

X_test=car_test.drop(columns=['name','selling_price',"owner","seats","torque"],axis=1)
y_test=car_test['selling_price']

X_train=car_train.drop(columns=['name','selling_price',"owner","seats","torque"],axis=1)
y_train=car_train['selling_price']
```

```python
X_test
```

# 2. Creating and Training the Model¶


```python
from sklearn.linear_model import LinearRegression

lm=LinearRegression()

lm.fit(X_train,y_train)
```

```python
print(lm.intercept_)
```

```python
temp=car.drop(columns=['name','selling_price',"owner","seats","torque"])
temp.head()
```

```python
coeff_df = pd.DataFrame(lm.coef_,temp.columns,columns=['Coefficient'])
coeff_df 
```

# 4. Predictions from our Model

```python
predictions=lm.predict(X_test)


fig = plt.figure()

fig.suptitle('y_test vs predictions')


plt.xlabel('y_test')


plt.ylabel('predcitions')
plt.scatter(y_test,predictions)
```

```python
fig=plt.figure(figsize=(8,8))
  
sns.distplot((y_test-predictions),bins=20)

fig.suptitle('Residual Analysis', fontsize = 20) 
```

# 5. Regression Evaluation Metrics

```python
from sklearn import metrics

print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

```python
R2 = metrics.r2_score(y_test,predictions)
R2
```
